```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from torch.autograd import Variable
from PIL import Image
```

```python
def preprocess(img):
  scaler = transforms.Resize((224,224))
  normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                   std=[0.229, 0.224, 0.225])
  to_tensor = transforms.ToTensor()
  t_img = Variable(normalize(to_tensor(scaler(img))).unsqueeze(0))
  return t_img
```

```python
resnet50 = models.resnet50(pretrained=True)
modules=list(resnet50.children())[:-1]
resnet50=nn.Sequential(*modules)
#for p in resnet50.parameters():
   # p.requires_grad = True
```

```python
img = preprocess(Image.open('../Recipe1M/recipe1M_images_train/train/0/0/0/0/0000dfbc52.jpg'))
preprocess_image = []
preprocess_image.append(img)
image_tensor = torch.Tensor(len(preprocess_image), 3, 224, 224)
torch.cat(preprocess_image, out=image_tensor)
image_tensor = torch.squeeze(resnet50(image_tensor), 2)
image_tensor = torch.squeeze(image_tensor, 2)
image_tensor.size()
```
