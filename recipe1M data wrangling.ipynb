```python
df_train = merged_json[merged_json['partition'] == "train"]
df_test = merged_json[merged_json['partition'] == "test"]
df_val = merged_json[merged_json['partition'] == "val"]
```

```python
del df_train['ingredients']
del df_train['url']
del df_train['partition']
del df_train['instructions']
del df_train['id']

del df_test['ingredients']
del df_test['url']
del df_test['partition']
del df_test['instructions']
del df_test['id']

del df_val['ingredients']
del df_val['url']
del df_val['partition']
del df_val['instructions']
del df_val['id']
```

```python
columns_titles = ["images","title"]
df_val=df_val.reindex(columns=columns_titles)
df_train=df_train.reindex(columns=columns_titles)
df_test=df_test.reindex(columns=columns_titles)
```

```python
df_val
```

```python
import json

val_json = json.loads(df_val.to_json(orient='records'))
train_json = json.loads(df_train.to_json(orient='records'))
test_json = json.loads(df_test.to_json(orient='records'))
```

```python
val_json
```

```python
out1 = open('./rec_imgID_train_new2.txt', 'w')
out2 = open('./rec_imgID_val_new2.txt', 'w')
out3 = open('./rec_imgID_test_new2.txt', 'w')

for data in val_json:
    out2.write(data['images'][0]['id'] + ' ' + data['title'] + '\n')
for data in train_json:
    out1.write(data['images'][0]['id'] + ' ' + data['title'] + '\n')
for data in test_json:
    out3.write(data['images'][0]['id'] + ' ' + data['title'] + '\n')
    
out1.close()
out2.close()
out3.close()
```
